- Finding the best straight line through data points
- The algorithm chooses the line that minimizes prediction errors called residuals
- Residual = actual - predicted

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area(sqr ft)") #feature (independent variable)
plt.ylabel("price(US $)") #target (dependent variable)
plt.scatter(df.area, df.price, color="blue", marker="*")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

- Fitting -> training the linear regression model using available data points
- First data point has to be a 2D array in ref.fit()

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("area") #feature (independent variable)
plt.ylabel("price") #target (dependent variable)
plt.scatter(df.area, df.price, color="blue", marker="*")
plt.plot(df.area,reg.predict(df[["area"]]),color='red')
```

- X must be 2D
- wrong ways -> reg.predict(3300), reg.predict([3300])
- correct -> reg.predict([[3300]])
- Because sklearn expects X (the features) to be a 2-dimensional structure, (number of samples, number of features)
