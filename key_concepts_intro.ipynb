# General Overview

- Supervised machine learning: Predict target values from labelled data (X->Y)
- Classification -> target values are discrete classes
- Regression -> target values are continuous classes

- Unsupervised machine learning: Find structure in unlabeled data
- Find groups of similar instances in the data (clustering)
- Finding unusual patterns (outlier detection)
- Ex - Flagging abnormal requests to server


1. Representation -> 2. Evaluation -> 3. Optimation -> Repeat 2 and 3
- Representation: Choose a feature representation and a type of classifier to use
- Evaluation: Choose which criterion distinguishes good vs bad classifiers
- Optimization: Choos how to search for settings/parameters that give the best classifier for this evaluation criterion

_A good classfier will have high accuracy so its prediction matches the true label a high % of the time_

**More Detail**  
- Convert each input object/sample into a set of features that describe the object
